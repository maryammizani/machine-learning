# 1. Load and arrange data into feature matrix and target vector

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris(as_frame=True)
```

```python
X = data.data
y = data.target
```

## Split arrays/matrices into random train and test subsets
Usually 80% of the data will be considered as training data and 20% as the test data.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=233)
```

```python
X_train.shape
```

```python
X_test.shape
```

# 2. Choose the model class

```python
from sklearn.linear_model import LogisticRegression
```

# 3. Instantiate the model

```python
logreg = LogisticRegression()
```

# 4. Fit the model 
Using the training data matrix and target array (80% of the data)

```python
logreg.fit(X_train, y_train)
```

# 5. Predict values for new data
Using the test data (20% of the data that was originally splitted)

```python
y_pred = logreg.predict(X_test)
```

```python
y_pred
```

## Evaluate the results
Compare the predicted values with the test target values (20% of the data that was originally splitted)

```python
y_test.values
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
